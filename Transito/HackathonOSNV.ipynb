```python
import io
import requests
import pandas as pd
```

```python


# URL of the Excel file
url = 'https://www.onsv.gob.pe/estaticos/excel/BBDD%20ONSV%20-%20SINIESTROS%202021-2022.xlsx'

# Download the file into memory
response = requests.get(url)
file_content = io.BytesIO(response.content)

# Load the Excel file starting from row 4
df = pd.read_excel(file_content, skiprows=3)

# Display the first few rows of the dataframe
df.head()
```
